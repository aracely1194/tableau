```python
import pandas as pd
```

```python
## Read in all files
data_jan = pd.read_csv("data/raw_datasets/JC-201901-citibike-tripdata.csv")
data_feb = pd.read_csv("data/raw_datasets/JC-201902-citibike-tripdata.csv")
data_mar = pd.read_csv("data/raw_datasets/JC-201903-citibike-tripdata.csv")
data_apr = pd.read_csv("data/raw_datasets/JC-201904-citibike-tripdata.csv")
data_may = pd.read_csv("data/raw_datasets/JC-201905-citibike-tripdata.csv")
data_jun = pd.read_csv("data/raw_datasets/JC-201906-citibike-tripdata.csv")
data_jul = pd.read_csv("data/raw_datasets/JC-201907-citibike-tripdata.csv")
data_aug = pd.read_csv("data/raw_datasets/JC-201908-citibike-tripdata.csv")
data_sep = pd.read_csv("data/raw_datasets/JC-201909-citibike-tripdata.csv")
data_oct = pd.read_csv("data/raw_datasets/JC-201910-citibike-tripdata.csv")
data_nov = pd.read_csv("data/raw_datasets/JC-201911-citibike-tripdata.csv")
data_dec = pd.read_csv("data/raw_datasets/JC-201912-citibike-tripdata.csv")
```

```python
## Add month column to each dataset
data_jan["month"]="Jan"
data_feb["month"]="Feb"
data_mar["month"]="Mar"
data_apr["month"]="Apr"
data_may["month"]="May"
data_jun["month"]="Jun"
data_jul["month"]="Jul"
data_aug["month"]="Aug"
data_sep["month"]="Sep"
data_oct["month"]="Oct"
data_nov["month"]="Nov"
data_dec["month"]="Dec"
```

```python
## Concat all dataframes into one
frames = [data_jan, data_feb, data_mar, data_apr, data_may, data_jun,
         data_jul, data_aug, data_sep, data_oct, data_nov, data_dec]

result = pd.concat(frames)
```

```python
## Rename gender to Unknown, Male and Female
result["gender"].replace({0: "Unknown", 1: "Male", 2: "Female"}, inplace=True)
result.head()
```

```python
result.to_csv('data/JC_citi_bike_2019.csv', index=False)
```
